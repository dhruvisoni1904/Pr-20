```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('ai-adoption-fortune500-synthetic-dataset-2020-2025.csv')

# Inspection
print("First 5 entries for AI Maturity Score:")
print(df['AI_Maturity_Score'].head())
```

```python
mean_val = df['AI_Maturity_Score'].mean()
median_val = df['AI_Maturity_Score'].median()
mode_val = df['AI_Maturity_Score'].mode()[0]

print(f"Mean: {mean_val:.2f}, Median: {median_val}, Mode: {mode_val}")
```

```python
range_val = df['AI_Maturity_Score'].max() - df['AI_Maturity_Score'].min()
variance_val = df['AI_Maturity_Score'].var()
std_dev = df['AI_Maturity_Score'].std()

print(f"Range: {range_val}, Std Dev: {std_dev:.2f}, Variance: {variance_val:.2f}")
```

```python
quartiles = df['AI_Maturity_Score'].quantile([0.25, 0.5, 0.75])
print(quartiles)
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['AI_Maturity_Score'], color='lightgreen')
plt.title('Statistical Spread of AI Maturity Scores')
plt.xlabel('AI Maturity Score')
plt.show()
```
